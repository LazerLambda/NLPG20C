```python
def load_reviews(path) -> list:
    import json
    reviews = list()

    with open(path, 'r', encoding="ISO-8859-1") as file:
        while file.readline():
            try:
                reviews.append(json.loads(file.readline()))
            except:
                continue

    return reviews
```

Import lib and load data

```python

import collections
import spacy
import en_core_web_sm
from spacy import displacy
reviews = load_reviews('reviewSelected100.json')
```

Init Spacy

```python
nlp = en_core_web_sm.load()
```

```python
pairs = list()

for e in reviews:
    text = e['text']
    doc = nlp(text)
    for token in doc:
        if token.dep_ == 'amod':
            pair = (( token.text.lower(), token.head.text.lower() ), (e['stars'],e['business_id'], e['useful'], e['funny'], e['cool']))
            pairs.append(pair)

observations = collections.defaultdict(int)

for e in pairs:
    if e[1][1] == 'vMpJzMFst_9GP4boeqWIRg':
        elem = e[0]
        if observations[elem]:
            observations[elem] += 1
        else:
            observations[elem] = 1
```
