```python
def load_reviews(path) -> list:
    import json
    reviews = list()

    with open(path, 'r', encoding="ISO-8859-1") as file:
        while file.readline():
            try:
                reviews.append(json.loads(file.readline()))
            except:
                continue

    return reviews
```

Import lib and load data

```python

import collections
import spacy
import en_core_web_sm
from spacy import displacy
reviews = load_reviews('reviewSelected100.json')
```

Init Spacy

```python
nlp = en_core_web_sm.load()
```

```python
pairs = list()

for e in reviews:
    text = e['text']
    doc = nlp(text)
    for token in doc:

#        if not e['business_id'] == 'vMpJzMFst_9GP4boeqWIRg':
#            continue

        # <Adj, Noun>
        pair = None

        # the case that the adjective is a direct amod
        if token.dep_ == 'amod':
            pair = ( token.text.lower(), token.head.text.lower() )
            pairs.append(pair)

        # the case that the adjective is a adj complement
        if token.dep_ == 'acomp':
            noun = [e for e in token.head.children if e.dep_ == "nsubj"]
            if noun:
                pair = ( token.text.lower(), noun[0].text.lower() )
                pairs.append(pair)
        
        # maybe using the lemma ?
        
observations = collections.Counter(pairs)#collections.defaultdict(int)

observations.most_common(20)
```
