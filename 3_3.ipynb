```python
def load_reviews(path) -> list:
    import json
    reviews = list()

    with open(path, 'r', encoding="ISO-8859-1") as file:
        while file.readline():
            try:
                reviews.append(json.loads(file.readline()))
            except:
                continue

    return reviews
```

```python
reviews = load_reviews('reviewSelected100.json')
```
